```python
import math
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

Cross entropy accross shifted Gaussian distributions

```python
# Set-up

sigma, mu_1, mu_2, mu_3 = 1, 0, 1, 3

x = np.linspace(-4, 8, 1000)

pdf_1 = norm.pdf(x, mu_1, sigma)
pdf_2 = norm.pdf(x, mu_2, sigma)
pdf_3 = norm.pdf(x, mu_3, sigma)
```

```python
# Plot of the distributions

plt.plot(x, pdf_1, label = 'pdf_1 & pdf_2')
plt.plot(x, pdf_2, label = 'pdf_1 & pdf_3')
plt.plot(x, pdf_3, label = 'pdf_2 & pdf_3')
plt.title( 'Normal Distribution PDFs')
plt.xlabel('x')
plt.ylabel ('Probability Density')
plt.legend()
plt.show()
```

```python
# Plots of cross entropy

H_12 = []
H_13 = []
H_23 = []

for i in range(len(pdf_1)):
    H_12.append(- pdf_1[i] * math.log(pdf_2[i]))
    H_13.append(- pdf_1[i] * math.log(pdf_3[i]))
    H_23.append(- pdf_2[i] * math.log(pdf_3[i]))

plt.plot(x, H_12, label = 'pdf_1 & pdf_2')
plt.plot(x, H_13, label = 'pdf_1 & pdf_3')
plt.plot(x, H_23, label = 'pdf_2 & pdf_3')
plt.legend()
plt.title( 'Cross Entropy across Shifted Distributions')
plt.xlabel('x')
plt.ylabel ('Probability Density')
plt.show()
```

Cross entropy accross Gaussian distributions with different spread

```python
mu, sigma_1, sigma_2, sigma_3 = 0, 1, 2, 6

x = np.linspace(-10, 10, 1000)

pdf_4 = norm.pdf(x, mu, sigma_1)
pdf_5 = norm.pdf(x, mu, sigma_2)
pdf_6 = norm.pdf(x, mu, sigma_3)

H_45 = []
H_46 = []
H_56 = []

for i in range(len(pdf_1)):
    H_45.append(- pdf_4[i] * math.log(pdf_5[i]))
    H_46.append(- pdf_4[i] * math.log(pdf_6[i]))
    H_56.append(- pdf_5[i] * math.log(pdf_6[i]))

plt.plot(x, H_45, label = 'pdf_4 & pdf_5')
plt.plot(x, H_46, label = 'pdf_4 & pdf_6')
plt.plot(x, H_56, label = 'pdf_5 & pdf_6')
plt.legend()
plt.title( 'Cross Entropy across Distributions Spread Differently')
plt.xlabel('x')
plt.ylabel ('Probability Density')
plt.show()
```
